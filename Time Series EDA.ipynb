## Datetime index

```python
data.info()
```

```python
data  = data.reset_index()
data
```

```python
pd.to_datetime(data['Date'])
```

```python
data = data.set_index('Date' , drop = True)
```

```python
data
```

## Time Resampling

```python
data.resample(rule = 'A').min()
```

```python
## Year-end frequency
data.resample(rule = 'A').max()['Open'].plot(color = 'purple')
```

```python
## Quarterly info
data.resample(rule = 'QS').max()['High'].plot()
```

```python
data['Open'].resample(rule = 'BA').mean().plot(kind = 'bar')
```

```python
data['Rolling_High_10'] = data['High'].rolling(window = 10, min_periods = 1).mean()
```

```python
data = data.drop('Rolling_High', axis = 1)
data.head(20)
```

```python
data[['High', 'Rolling_High_10']].plot(figsize = (12,6), xlim = ['2020-01-01', '2023-02-01'])
```

```python
data['Rolling_Open_10'] = data['Open'].rolling(window = 10, min_periods = 1).mean()
data['Rolling_Open_30'] = data['Open'].rolling(window = 30, min_periods = 1).mean()
data['Rolling_Open_45'] = data['Open'].rolling(window = 45, min_periods = 1).mean()
```

```python
data[['Open','Rolling_Open_10', 'Rolling_Open_30', 'Rolling_Open_45']].plot(figsize = (12,6), xlim = ['2020-01-01', '2023-02-01'])
```

```python
data['Open_expanding']= data['Open'].expanding().mean()
```

```python
data.head()
```

## Exapnding Features

```python
data['Open_expanding'].plot(figsize = (12,6))
```

## Exponential Weighted Moving Average

```python
data['Open_EMA'] = data['Open'].ewm(alpha = 0.1, adjust=False).mean()
```

```python
data[['Open', 'Open_EMA']].plot(figsize = (12,6),xlim = ['2020-01-01', '2023-02-01'])
```

```python
data['Open_EMA2'] = data['Open'].ewm(alpha = 0.3, adjust=False).mean()
```

```python
data[['Open', 'Open_EMA', 'Open_EMA2']].plot(figsize = (15,6),xlim = ['2022-01-01', '2023-02-01'])
```
