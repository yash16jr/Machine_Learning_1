# Outlier Treatment

```python
df.info()
```

```python
uv = np.percentile(df.n_hot_rooms, [99])[0] 
# gives the value of 99th percentile of the data
# we want zeroeth index value as there may be several values in 99 percentile
```

```python
uv
```

```python
df[(df.n_hot_rooms > uv)]
# this returns the rows in which the value of nhotrooms > uv
```

```python
#Capping

df.n_hot_rooms[(df.n_hot_rooms> 3*uv)] = 3*uv

# the value of outliers on the upper side gets replaced by 3 times of upper limit 
```

```python
sns.jointplot(x= 'n_hot_rooms' , y = 'price',data = df)
```

```python
lv = np.percentile(df.rainfall, [1])[0]
lv
```

```python
df[(df.rainfall < lv)]
```

```python
df.rainfall[(df.rainfall < 0.3*lv)] = 0.3*lv
```

```python
sns.jointplot(x= 'crime_rate', y = 'price', data = df )
```

# Decision Trees

```python
df3 = pd.read_csv('C:/Users/Yash/Downloads/Data Files/3. ST Academy - Decision Trees resource files/Movie_regression.csv' , header = 0)
```

```python
df3.head()
```

```python
df3.info()
```

## Missing Value Imputation

```python
df3['Time_taken'].fillna(value = df3['Time_taken'].mean() , inplace = True )
```

## Dummy Variable Creation
### Conversion of categorical values to numerical values

```python
df3.head()
```

```python
df3 = pd.get_dummies(df3, columns = ["3D_available" ,"Genre"] , drop_first = True)
```

```python
df3.head()
```

## X-y Split

```python
Xdt = df3.loc[:, df3.columns != 'Collection']
## x of decison tree
```

```python
ydt = df3['Collection']
```

## Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
Xdt_train, Xdt_test, ydt_train, ydt_test = train_test_split(Xdt,ydt,test_size = 0.2, random_state = 0)
```

## Training Regression tree

```python
from sklearn import tree
regtree = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
regtree.fit(Xdt_train, ydt_train)
```

## Predicting Values 

```python
ydt_train_pred = regtree.predict(Xdt_train)
ydt_test_pred = regtree.predict(Xdt_test)
```

## Checking performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(ydt_test,ydt_test_pred)
```

```python
r2_score(ydt_train,ydt_train_pred)
```

```python
r2_score(ydt_test,ydt_test_pred)
```

```python
dftest = pd.DataFrame({'Actual' : ydt_test,'Predicted' : ydt_test_pred})
dftest
```

## Plotting Decision Tree

```python
dot_data = tree.export_graphviz(regtree, out_file = None, feature_names = Xdt_train.columns, filled = True)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

## Controlling the tree growth

## Minimum observations at internal Node Method

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split = 40)
regtree2.fit(Xdt_train,ydt_train)
dot_data = tree.export_graphviz(regtree2, out_file = None, feature_names = Xdt_train.columns, filled = True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

## Minimum obseravtions at leaf nodes

```python
regtree3 = tree.DecisionTreeRegressor(min_samples_leaf = 25 )
regtree3.fit(Xdt_train, ydt_train)
dot_data = tree.export_graphviz(regtree3, out_file = None, feature_names = Xdt_train.columns, filled = True)
graph3 = pydotplus.graph_from_dot_data(dot_data)
Image(graph3.create_png())
```
